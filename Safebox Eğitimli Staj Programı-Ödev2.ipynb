```python
import json
import os

def film_ekle():
    film_adı = input("Film adını girin: ")
    film_türü = input("Film türünü girin: ")
    film_tarihi = input("Film tarihini girin: ")
    film_yönetmeni = input("Film yönetmenini girin: ")

    with open("movies.json", "r") as file:
        movies = json.load(file)

    movies.append({"ad": film_adı, "tür": film_türü, "tarih": film_tarihi, "yönetmen": film_yönetmeni})

    with open("movies.json", "w") as file:
        json.dump(movies, file)

    print("Film başarıyla eklendi.")

def film_sil():
    film_adı = input("Silmek istediğiniz film adını girin: ")

    with open("movies.json", "r") as file:
        movies = json.load(file)

    updated_movies = [movie for movie in movies if movie["ad"] != film_adı]

    with open("movies.json", "w") as file:
        json.dump(updated_movies, file)

    print("Film başarıyla silindi.")

def film_ara():
    film_adı = input("Aramak istediğiniz film adını girin: ")

    with open("movies.json", "r") as file:
        movies = json.load(file)

    found_movies = [movie for movie in movies if film_adı.lower() in movie["ad"].lower()]

    if found_movies:
        print("Bulunan filmler:")
        for movie in found_movies:
            print(f"Ad: {movie['ad']}, Tür: {movie['tür']}, Tarih: {movie['tarih']}, Yönetmen: {movie['yönetmen']}")
    else:
        print("Film bulunamadı.")

def cikis():
    print("Programdan çıkılıyor.")
    quit()

def menu_goster():
    print("**********")
    print("Film Veritabanı")
    print("**********")
    print("1. Film Ekle")
    print("2. Film Sil")
    print("3. Film Ara")
    print("4. Çıkış")
    print("**********")

# movies.json dosyasını oluştur
if not os.path.exists("movies.json"):
    with open("movies.json", "w") as file:
        json.dump([], file)

menu_goster()

while True:
    secim = input("Yapmak istediğiniz işlemi seçin (1-4): ")

    if secim == "1":
        film_ekle()
    elif secim == "2":
        film_sil()
    elif secim == "3":
        film_ara()
    elif secim == "4":
        cikis()
    else:
        print("Geçersiz bir seçim yaptınız. Tekrar deneyin.")
```
